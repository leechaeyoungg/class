```python
url='https://www.gutenberg.org/cache/epub/11/pg11-images.html'
driver.get(url)
```

```python
html=driver.page_source
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,'html.parser')
soup
```

```python
alice_content=soup.select('.chapter > p') #본문만 선택
alice_content
```

```python
alice_content[0].text
```

```python
alice=[]
for i in range(len(alice_content) ):
    ac = alice_content[i].text.replace('\n','')
    alice.append(ac)
```

```python
alice
```

```python
import re
import urllib.request
import zipfile
from lxml import etree
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
alice=str(alice)
alice=alice.strip('[')
alice=alice.strip(']')
alice=alice.strip("'")
alice
```

```python
sent_text = sent_tokenize(alice)
```

```python
sent_text
```

```python
normalized_text = []
for string in sent_text:
     tokens = re.sub(r"[^a-z0-9]+", " ", string.lower())
     normalized_text.append(tokens)
```

```python
normalized_text
```

```python
result = [word_tokenize(sentence) for sentence in normalized_text]
```

```python
result
```

```python
len(result) 
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
```

```python
model=Word2Vec(sentences=result,vector_size=100,window=5,min_count=5,sg=0,workers=4)
```

```python
model.wv.most_similar('rabbit')
```

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
tokens = word_tokenize(alice)
pos_tags = nltk.pos_tag(tokens)

```

```python
filtered_words = []

for word, pos in pos_tags:
    if pos.startswith('N')  or pos.startswith('J'):
        filtered_words.append(word)

filtered_text = ' '.join(filtered_words)
```

```python
filtered_text
```

```python
sent_text = sent_tokenize(filtered_text)
```

```python
sent_text
```

```python
normalized_text = []
for string in sent_text:
     tokens = re.sub(r"[^a-z0-9]+", " ", string.lower())
     normalized_text.append(tokens)
```

```python
normalized_text
```

```python
result = [word_tokenize(sentence) for sentence in normalized_text]
```

```python
result
```

```python
model=Word2Vec(sentences=result,vector_size=10,window=6,min_count=5,sg=0,workers=12)
```

```python
model.wv.most_similar('cat')
```

```python
model.wv.most_similar('king')
```

```python
model.wv.save_word2vec_format('alice_w2v') #모델 저장
```

```python
loaded_model=KeyedVectors.load_word2vec_format('alice_w2v')
#불러오기
```

```python
loaded_model.most_similar('king')
```
