```python
parse_text = '\n'.join(target_text.xpath("//content/text()"))
```

```python
#소괄호로 묶여있는 부분 빼기 (Audio),...
```

```python
content_text = re.sub(r'\([^)]*\)', '', parse_text)
#소괄호로 묶인 내용 전부 삭제
```

```python
import nltk
nltk.download('punkt')
```

```python
sent_text = sent_tokenize(content_text)
```

```python
sent_text
```

```python
normalized_text = []
for string in sent_text:
     tokens = re.sub(r"[^a-z0-9]+", " ", string.lower())
     normalized_text.append(tokens)
```

```python
normalized_text
```

```python
result = [word_tokenize(sentence) for sentence in normalized_text]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Okt
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt")
#네이버 영화 리뷰 긍정 부정
```

```python
train_data = pd.read_table('ratings.txt')
```

```python
train_data
```

```python
#document열에 있는 단어들 word to vector 하기
```

```python
train_data = train_data.dropna(how = 'any')
#null값 제거
```

```python
print(len(train_data))
```

```python
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
#한글과 공백문자 제외 나머지 전부 제거
```

```python
train_data[:5]
```

```python
# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

# 형태소 분석기 OKT를 사용한 토큰화 작업 (다소 시간 소요)
okt = Okt()
```

```python
#konlpy.org -> api -> okt -> 함수
```

```python
train_data['document'][0]
```

```python
okt.morphs(train_data['document'][0])
```

```python
okt.morphs(train_data['document'][0],stem=True)
#어간추출
```

```python
pip install tqdm
```

```python
from tqdm import tqdm
```

```python
#tqdm -> 진행상황 보여줌
```

```python
tokenized_data = []
for sentence in tqdm(train_data['document']):
  #어릴때보고 지금다시봐도 재밌어요ㅋㅋ
    tokenized_sentence = okt.morphs(sentence, stem=True) # 토큰화
    ['어리다', '때', '보고', '지금', '다시', '보다', '재밌다', 'ㅋㅋ']
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    tokenized_data.append(stopwords_removed_sentence)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences = tokenized_data, vector_size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
```

```python
model.wv.vectors.shape
#word to vector의 크기
```

```python
print(model.wv.most_similar("쓰레기"))
```

```python
print(model.wv.most_similar("싸이코"))
```

```python
print(model.wv.most_similar("초승달"))
#5번 미만이면 안나옴 (min_count = 5로 설정해서)
```
