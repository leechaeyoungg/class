```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1) #
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
x_train.shape
y_train.shape
```

```python
#  
```

```python
W=torch.zeros(1,requires_grad=True)
W
#   
# requires_grad=True  ->W  (   )
```

```python
b=torch.zeros(1,requires_grad=True) #b ->  
```

```python
#   : y = 0 * x + 0
```

```python
#  
# y = 0 * x + 0   
hx=W * x_train + b
```

```python
hx
# 0 ()
```

```python
#  hx  
```

```python
#cost 

cost=torch.mean((hx - y_train)**2)
cost
```

```python
optimizer=optim.SGD([W,b],lr=0.01)
```

```python
optimizer.zero_grad() #    0 
cost.backward()      # . w, b    
optimizer.step()    # W,b  
```

```python
for epoch in range(2000):
    hx=W * x_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer=optim.SGD([W,b],lr=0.01) #  
    optimizer.zero_grad()#    0 
    cost.backward() #W, b   
    optimizer.step()#W,b  
    if epoch%100 ==0:
        print(epoch,  W.item(), b.item(), cost.item())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
model=nn.Linear(1,1) # ,
```

```python
list(model.parameters()) #w b  
# : y=0.5153 + x_train -0.4414
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(2000):
    
    hx = model(x_train) #hx=x_train.matmul(W) + b
    cost = F.mse_loss(hx, y_train)   #cost=torch.mean((hx - y_train)**2)
    
    
    optimizer.zero_grad()#    0 
    cost.backward() #W, b   
    optimizer.step()#W,b  
    if epoch%100 ==0:
        print(epoch , cost.item())
```
