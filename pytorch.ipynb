```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1) #시드설정
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
x_train.shape
y_train.shape
```

```python
#가중치 먼저 정의
```

```python
W=torch.zeros(1,requires_grad=True)
W
# 초기값 임의로 줘도됨
# requires_grad=True  ->W는 변수 (학습하는 과정에서 업데이트되는 대상)
```

```python
b=torch.zeros(1,requires_grad=True) #b -> 도 변수
```

```python
#초기 모델의 가설함수 : y = 0 * x + 0
```

```python
#초기 모델 가설함수
# y = 0 * x + 0 를 그대로 작성
hx=W * x_train + b
```

```python
hx
#다 0으로 예측(초기모델이므로)
```

```python
#훈련하는 과정에서 hx가 달라질 것
```

```python
#cost 정의

cost=torch.mean((hx - y_train)**2)
cost
```

```python
optimizer=optim.SGD([W,b],lr=0.01)
```

```python
optimizer.zero_grad() #미분을 해서 구한 기울기를 0으로 초기화해줌
cost.backward()      #역전파 함수. w, b 에 대한 기울기 계산
optimizer.step()    # W,b에 대한 업데이트
```

```python
for epoch in range(2000):
    hx=W * x_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer=optim.SGD([W,b],lr=0.01) #바깥에 있어도 됨
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch,  W.item(), b.item(), cost.item())
```

```python
x1_train = torch.FloatTensor([[73], [93], [89], [96], [73]])
x2_train = torch.FloatTensor([[80], [88], [91], [98], [66]])
x3_train = torch.FloatTensor([[75], [93], [90], [100], [70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
#수능점수 예측하는 모델을 만들기
```

```python
x1_train.shape #5명의 데이터가 있다고 생각
```

```python
optimizer=optim.SGD([w1, w2, w3, b],lr=0.00001) # lr learning rate hyper parameter 0.01 <= lr <= 0.001
# 1* 10의 -5승 = 1* 0.0001 이므로 1e-5로 표현해도됨

for epoch in range(2000):
    hx=w1 * x1_train + w2 * x2_train + w3 * x3_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%10 ==0:
        print(epoch,  w1.item(),  w2.item(),  w3.item(), b.item(), cost.item())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
model=nn.Linear(1,1) #입력차원 ,출력차원
```

```python
list(model.parameters()) #w하고 b값이 랜덤하게 설정
#초기 모델: y=0.5153 + x_train -0.4414
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(2000):
    
    hx = model(x_train) #hx=x_train.matmul(W) + b
    cost = F.mse_loss(hx, y_train)   #cost=torch.mean((hx - y_train)**2)
    
    
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch , cost.item())
```
